```python
from bs4 import BeautifulSoup
import urllib2
import os
```

```python
direc = "./Train/DrugBank/"
files = os.listdir(direc)
new_sent = []
for f in files:
    print f
    content = open(direc + f, "r")
    soup = BeautifulSoup(content,  "lxml")
    sentences = soup.find_all("sentence")
    for sentence in sentences:
        text = sentence["text"]
        entities = sentence.find_all("entity")
        entity_dict = {}
        for entity in entities:
            drug_id = entity["id"]
            drug_name = entity["text"]
            entity_dict[drug_id] = drug_name
        pairs = sentence.find_all("pair")
        for pair in pairs:
            e1 = entity_dict[pair["e1"]]
            e2 = entity_dict[pair["e2"]]
            if e1 == e2:
                continue
            ddi = pair["ddi"]
            if ddi == "true":
                ddi_type = pair["type"]
                copy_text = text
                copy_text = copy_text.replace(e1, "<e1>" + e1 + "</e1>", 1)
                copy_text = copy_text.replace(e2, "<e2>" + e2 + "</e2>", 1)
                copy_text = copy_text.replace("\n", "")
                op = copy_text + "\t" + ddi_type + "(e1,e2)\n"
                if op not in new_sent:
                    new_sent.append(op)
            else:
                ddi_type = "None"
                copy_text = text
                copy_text = copy_text.replace(e1, "<e1>" + e1 + "</e1>", 1)
                copy_text = copy_text.replace(e2, "<e2>" + e2 + "</e2>", 1)
                copy_text = copy_text.replace("\n", "")
                op = copy_text + "\t" + ddi_type + "(e1,e2)\n"
                if op not in new_sent:
                    new_sent.append(op)
    content.close()
target = open("./Train/DrugBank.txt", "w")
for sent in new_sent:
    target.write(sent)
target.close()
```

```python
direc = "./Train/MedLine/"
files = os.listdir(direc)
new_sent = []
for f in files:
    print f
    content = open(direc + f, "r")
    soup = BeautifulSoup(content,  "lxml")
    sentences = soup.find_all("sentence")
    for sentence in sentences:
        text = sentence["text"]
        entities = sentence.find_all("entity")
        entity_dict = {}
        copy_text1 = text
        for entity in entities:
            drug_id = str(entity["id"])
            drug_name = str(entity["text"])
            entity_dict[drug_id] = drug_name
            copy_text1 = copy_text1.replace(drug_name, drug_id, 1)
        pairs = sentence.find_all("pair")
        for pair in pairs:
            e1 = pair["e1"]
            e2 = pair["e2"]
            ddi = pair["ddi"]
            if ddi == "true":
                copy_text = copy_text1
                ddi_type = pair["type"]
                copy_text = copy_text.replace(e1, "<e1>" + entity_dict[e1] + "</e1>", 1)
                copy_text = copy_text.replace(e2, "<e2>" + entity_dict[e2] + "</e2>", 1)
                for key in entity_dict:
                    if key != e1 and key !=e2:
                        copy_text = copy_text.replace(key, entity_dict[key], 1)
                op = copy_text + "\t" + ddi_type + "(e1,e2)"
                op = op.replace("\n", "")
                op = op.replace("\r", "")
                if op not in new_sent and "<e1>" in op and "<e2>" in op:
                    new_sent.append(op)
            else:
                copy_text = copy_text1
                ddi_type = "None"
                copy_text = copy_text.replace(e1, "<e1>" + entity_dict[e1] + "</e1>", 1)
                copy_text = copy_text.replace(e2, "<e2>" + entity_dict[e2] + "</e2>", 1)
                for key in entity_dict:
                    if key != e1 and key !=e2:
                        copy_text = copy_text.replace(key, entity_dict[key], 1)
                op = copy_text + "\t" + ddi_type + "(e1,e2)"
                op = op.replace("\n", " ")
                if op not in new_sent and "<e1>" in op and "<e2>" in op:
                    new_sent.append(op)
                
    content.close()
target = open("./Train/Medline.txt", "w")
for sent in new_sent:
    target.write(sent)
    target.write("\n")
target.close()
```
